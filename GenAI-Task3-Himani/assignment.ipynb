## TASK 1

```python
def apply_discount(amount, discount_percentage=5):
    if discount_percentage > 60:
        return "discount percentage cannot be more than 60%"
    return amount - amount*discount_percentage/100

print(apply_discount(1000,10))
print(apply_discount(500))
print(apply_discount(1000,70))
```

## TASK 2

```python
def factorial(n):
    if n<0:
        return "Cannot calculate factorial of negative numbers"
    elif n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
print(factorial(0))
print(factorial(-3))
        
```

## TASK 3

```python
gst= lambda price: price*0.18 + price
print(gst(1000))
discounted_price_with_gst = lambda price, discount=5: price - (price*discount/100) + (price - (price*discount/100))*0.18
print(discounted_price_with_gst(1000,10)) 
```

```python
prices = [100, 250, 400, 1200, 50]
price_with_gst=list(map(gst,prices))
print('original prices: ', prices)
print('prices after gst: ', price_with_gst)
```

## TASK 5

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]
expensive_items = list(filter(lambda price: price > 500, prices))
non_expensive_items = list(filter(lambda price: price <= 500, prices))
print('expensive items: ', expensive_items)
print('non-expensive items: ', non_expensive_items)
```

## TASK 6

```python
def process_prices(prices):
    discounted_prices=list(map(lambda price: price - price*0.1, prices))
    filtered_list=list(filter(lambda price: price > 300, discounted_prices))
    return discounted_prices, filtered_list
prices = [100, 500, 900, 50, 750]
discounted_prices, filtered_list = process_prices(prices)
print('discounted prices: ', discounted_prices)
print('filtered list: ', filtered_list)
```

## TASK 7

```python
add_price = lambda price_list, price: price_list.append(price)
get_average_price = lambda price_list: sum(price_list)/len(price_list)
get_max_price = lambda price_list: max(price_list)
prices = [100, 250, 400]

while True:
    option = input('Choose an option: \n1 add a price to list \n2 get average price \n3 get maximum price \nq exit\n')
    if option == '1':
        price = float(input('Enter the price to add: '))
        add_price(prices, price)
        print(f'Price {price} added to list.')
    elif option == '2':
        average_price = get_average_price(prices)
        print(f'Average price: {average_price}')
    elif option == '3':
        max_price = get_max_price(prices)
        print(f'Maximum price: {max_price}')
    elif option.lower() == 'q':
        print('Exiting...')
        break
    else:
        print('Invalid option. Please try again.')
```
