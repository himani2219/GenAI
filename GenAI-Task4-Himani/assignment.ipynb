## TASK 1

```python
# sales list
sales=[1200, 450, 980, 1500, 3000]
# write sales data to a text file
f=open('sales_data.txt', 'w')
for sale in sales:
    f.write(str(sale) + '\n')
f.close()
# read from the text file
f=open('sales_data.txt', 'r')
print(f.read())
f.close()
# OPTIONAL TASK : to write as comma separated values
f=open('sales_data.csv', 'w')
for sale in sales:
    f.write(str(sale) + ',')
f.close()
# to read from the csv file 
f=open('sales_data.csv', 'r')
print(f.read())
f.close()
```

## TASK 2

```python
f = open('sales_data.txt', 'r')
sales_data = f.read()
f.seek(0)
sales_data_line1 = f.readline()
f.seek(0)
sales_list = f.readlines()
for i in range(len(sales_list)):
    sales_list[i]=int(sales_list[i].strip())
# for loop is used ot remove newline character and convert str to int
print('sales_data: ', sales_data)
print('sales_data_line1: ', sales_data_line1)
print('sales_list: ', sales_list)
f.close()

# need to seek the pointer to the start of the file or else it will read from current position and no output will be there 
```

## TASK 4

```python
with open('sales_data.txt', 'r') as f:
    # reading from file 
    print(f.read())
    # moving pointer to start of file 
    f.seek(0)
    sales_data = f.readlines()
    # cleaning data and converting to int 
    for i in range(len(sales_data)):
        sales_data[i] = int(sales_data[i].strip())
    # calculations 
    print('Total sales: ', sum(sales_data))
    print('Highest sale: ', max(sales_data))
    print('Lowest sale: ', min(sales_data))
    print('Average sale: ', sum(sales_data)/len(sales_data))
```

## TASK 5

```python
product_info = {}
# asking user to enter product name and price in comma separated format 
for i in range(3):
    product = input('Enter product name and price (comma separated): ')
    name, price = product.split(',')
    product_info[name.strip()]=float(price.strip())
# writing into file in the format "product name | price" 
with open('products.txt', 'w') as f:
    for name, price in product_info.items():
        f.write(f'{name} | {str(price)}\n')
# reading from file and printing product info
with open('products.txt', 'r') as f:
    print(f.read())
```

## TASK 7

```python
prices={"mouse": 500, "keyboard": 800, "monitor": 7000, "pendrive": 400, "camera": 5000}
discount_percentage = int(input('Enter discount percentage to apply: '))
with open('discount_report.txt', 'w') as f:
    total_discounted_price = 0
    for product, price in prices.items():
        # calculate discounted price 
        discounted_price = price - price*discount_percentage/100
        total_discounted_price += discounted_price
        f.write(f'{product} | original price: {price} | discounted price: {discounted_price}\n')
    # OPTIONAL: to write total items and total discounted price at the end of the file
    f.write('\nTotal items: ' + str(len(prices)))
    f.write('\nTotal discounted price: ' + str(total_discounted_price))

# open file and print
with open('discount_report.txt', 'r') as f:
    print(f.read())
```
