```python
# 1. dictionary in the format key=product name and value=prices
price_dict = {'notebook':100, 'laptop':100000, 'pen':10.50, 'shirt':3000, 'handbag':30000, 'watch':10000}
```

```python
# 2. add a new product, update the price of an existing product, delete a product
new_prod = {'bracelet': 400}
price_dict={**price_dict, **new_prod}
price_dict['laptop']=130000
if price_dict['pen']:
    del price_dict['pen']
else:
    print('item does not exist')
price_dict
```

```python
# 3. average price of all products
avg_price = sum(price_dict.values())/len(price_dict)
print(avg_price)
```

```python
# 4. print the value of max and min price values
min_item = min(price_dict, key=price_dict.get)
max_item = max(price_dict, key=price_dict.get)
print(f'the maximum price item is {max_item} with value {price_dict[max_item]}')
print(f'the minimum price item is {min_item} with value {price_dict[min_item]}')
```
