```python
product_list = ['pen', 'mouse', 'tshirt', 'keyboard', 'sharpener']
price_dict = {'notebook':100, 'laptop':100000, 'pen':10.50, 'tshirt':3000, 'keyboard':3000, 'sharpener':10.75, 'mouse':2000}
category_list = ['stationary', 'electronics', 'clothing', 'electronics', 'stationary']

catalog = []
for i in range(0,len(product_list)):
    catalog.append((product_list[i], price_dict[product_list[i]], category_list[i] ))
print(catalog)
```

```python
category_to_products={}
for (item, price, category) in catalog:
    if category not in category_to_products:
        category_to_products[category]=[]
    category_to_products[category].append(item)
print(category_to_products)
```

```python
max_items_category_products = max(category_to_products, key=lambda x: len(category_to_products[x]))
category_to_products[max_items_category_products]
```
