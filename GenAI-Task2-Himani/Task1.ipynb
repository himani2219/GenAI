program that reads an integer order_amount from user

```python
try:
    order_amount = int(input('Please enter order amount: '))
    print('Order amount is: ', order_amount)
except ValueError:
    print('Please enter a valid integer for order amount')
```

apply discount rules and final amount

```python
discount = 0
if(order_amount>=2000):
    discount = 0.15*order_amount
elif(order_amount>=1500): 
    discount = 0.1*order_amount
elif(order_amount>=1000):
    discount = 0.07*order_amount
total_amount = order_amount - discount
tax = 0.05*total_amount
net_payable_amount = total_amount + tax
print(f'total amount is: {order_amount}')
print(f'discount amount is: {discount}')
print(f'total amount after discount is: {total_amount}')
print(f'tax amount is: {tax}')
print(f'net payable amount is: {net_payable_amount}')
```
