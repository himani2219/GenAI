```python
def discounted_price(order_amount):
    discount = 0
    if(order_amount>=2000):
        discount = 0.15
    elif(order_amount>=1500): 
        discount = 0.1
    elif(order_amount>=1000):
        discount = 0.07
    return discount
```

```python
orders=[1200,2500,800,1750,3000]
for order in orders:
    print(f'Order amount: {order}, discount percentage: {discounted_price(order) * 100}%, net price: {order - (order * discounted_price(order))}')
```

```python
total_revenue = 0
for order in orders:
    total_revenue += order - (order * discounted_price(order))
print(f'Total revenue: {total_revenue}')
```
