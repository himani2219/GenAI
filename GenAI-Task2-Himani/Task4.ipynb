```python
daily=[200, 150, 0, 400, 50, -1, 300]
total_sales = 0
for sale in daily:
    if sale == -1:
        print("Corrupt data!!! Terminating further processing...")
        break
    elif sale == 0:
        print("No sales...")
        continue
    else:
        total_sales += sale
        print(f'Running sales: {total_sales}')
print(f'Total sales: {total_sales}')

```
