user driven menu

```python
orders=[1200,2500,800,1750,3000]
```

```python
while True:
    try:
        option = input('select an option frm the menu: \n1 Add an order amount to the list. \n2 show all orders and total after applying discount. \nq quit')
        if option == '1':
            order=int(input('enter an order amount: '))
            orders.append(order)
        elif option == '2':
            discount=0
            for order in orders:
                if order >= 2000:
                    discount=15
                elif order >= 1500:
                    discount=10
                elif order >= 1000:
                    discount=7
                else:
                    discount=0
                print(f'Order amount: {order}, discount percentage: {discount}%, net price: {order - discount/100*order}')
        elif option == 'q':
            break
        else:
            print('Invalid option. please try again...')
    except Exception as e:
        print(e, 'please try again...')
        continue

```
